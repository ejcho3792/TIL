(2) 기온 데이터 히스토그램

```python
import csv
import matplotlib.pyplot as plt

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))
        
plt.hist(result,bins=100,color='r')
plt.show()
```

(3) 특정 월의 기온 히스토그램

```python
import csv
import matplotlib.pyplot as plt

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
result=[]
# 특정할 월을 입력
my_month='08'

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month==my_month:
            result.append(float(row[-1]))
        
plt.hist(result,bins=100,color='r')
plt.show()
```

(4) 2개 월의 기온 히스토그램

```python
import csv
import matplotlib.pyplot as plt

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
res1=[]
res2=[]

# 특정할 월을 입력
my_month1='02'
my_month2='08'

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month==my_month1:
            res1.append(float(row[-1]))
        if month==my_month2:
            res2.append(float(row[-1])) 
            
plt.hist(res1,bins=100,color='b',label=my_month1)
plt.hist(res2,bins=100,color='r',label=my_month2)
plt.legend()
plt.show()
```

2. 박스플롯 ( boxplot() )

(1) 임의의 숫자를 이용한 박스플롯

```python
import matplotlib.pyplot as plt
import random
import numpy as np
result=[]
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
res=np.array(result)
print('1/4: '+str(np.percentile(res,25)))
print('2/4: '+str(np.percentile(res,50)))
print('3/4: '+str(np.percentile(res,75)))
```

(3) 기온 데이터의 월별 박스플롯

```python
import csv

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
result=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1]!='':
        result[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(result)
plt.show()
```

```python
import csv

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
result=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1]!='':
        result[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.style.use('ggplot')               # ggplot : 회색 격자무늬 배경에 박스 2/4 파란선 추가
plt.figure(figsize=(10,5), dpi=300)   # 그래프 크기 : 가로 10, 세로 5
plt.boxplot(result,showfliers=False) # showfliers = 아웃라이어 삭제
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
day=[]
for i in range(31):
    day.append([])
for row in data:
    if row[-1]!='':
        day[int(row[0].split('-')[2])-1].append(float(row[-1]))
        
plt.style.use('ggplot')               # ggplot : 회색 격자무늬 배경에 박스 2/4 파란선 추가
plt.figure(figsize=(10,5), dpi=300)   # 그래프 크기 : 가로 10, 세로 5
plt.boxplot(day,showfliers=False) # showfliers = 아웃라이어 삭제
plt.show()
```
