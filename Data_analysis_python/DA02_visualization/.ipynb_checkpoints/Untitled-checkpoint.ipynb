# Graph

---
### 1. matplotlib library

```python
import matplotlib.pyplot as plt
```

(1) Set legend (label= 'name')

```python
plt.title('legend')       # set title
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend()
plt.show()
```

(2) Graph color(color='color name')

```python
plt.title('color')       # set title
plt.plot([10,20,30,40],color='y',label='yellow')
plt.plot([40,30,20,10],color='pink',label='pink')
plt.legend()
plt.show()
```

(3) Line style(linestyle='line style' or ls='line style')

```python
plt.title('linestyle')       # set title
plt.plot([10,20,30,40],color='y',linestyle='--',label='dashed')
plt.plot([40,30,20,10],color='pink',ls=':',label='dotted')
plt.legend()
plt.show()
```

(4) Marker style 

```python
plt.title('marker')       # set title
plt.plot([10,20,30,40],'r.',label='circle')
plt.plot([40,30,20,10],'g^',label='triangle up')
plt.legend()
plt.show()
```

(5) x-y graph

```python
plt.plot([1,2,3,4],[12,43,25,15])           # plt.plot([x_data],[y_data])
plt.show
```
