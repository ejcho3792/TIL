# 특정 날짜의 기온 변화

```python
# import library
import csv
import matplotlib.pyplot as plt
```

```python
f=open('../data/seoul_new.csv')
data=csv.reader(f)
next(data)
high=[]                          # 최고 기온을 저장
low=[]                           # 최저 기온을 저장

my_date='19960404'               # 특정 날짜를 입력

for row in data:
    if row[-1]!='' and row[-2]!='':
        date=row[0].split('-')
        if int(my_date[:4])<=int(date[0]):
            if date[1]==my_date[4:6]and date[2]==my_date[6:]:
                high.append(float(row[-1]))
                low.append(float(row[-2]))
```

```python
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('특정일의 기온 변화 그래프')
plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend()
plt.show()
```
