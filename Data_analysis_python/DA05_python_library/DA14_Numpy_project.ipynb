## 데이터 불러오기

## 관심지역 입력받기

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
d_pth='../data/age.csv'
f=open(d_pth)
data=csv.reader(f)
header=next(data)
name=input('관심지역을 입력하세요 : ')
# row[0] 행정구역, row[1], row[2] : 총 인구수 
for row in data:
    if name in row[0]:
        for i in row[3:]:
            res=np.array(row[3:],dtype=int)
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300)
plt.style.use('ggplot')
plt.plot(res)
plt.show()
f.close()
```

## 관심지역과 가장 유사한 인구 구조를 가진 지역 찾기

```python
f=open(d_pth)
data=csv.reader(f)
header=next(data)
data=list(data)
name=input('관심지역을 입력하세요 : ')
mn=1
res=0
res_name=''
for row in data:
    row[1:]=[r.replace(',','') for r in row[1:]]
    if name in row[0]:
        for i in row[3:]:
            home=np.array(row[3:],dtype=int)/int(row[2])
            
for row in data:
    row[1:]=[r.replace(',','') for r in row[1:]]
    away=np.array(row[3:],dtype=int)/int(row[2])
    s=np.sum((home-away)**2)
    if s<mn and name not in row[0]:
        mn=s
        res_name=row[0]
        res=away
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300)
plt.style.use('ggplot')
plt.plot(home,label=name)
plt.plot(res,label=res_name)
plt.legend()
plt.show()
f.close()
```
