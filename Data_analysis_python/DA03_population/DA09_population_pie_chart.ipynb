## 제주도의 여성비율을 barh plot 하기

```python
import csv
import matplotlib.pyplot as plt

d_pth='../data/gender.csv'
```

```python
f=open(d_pth)
data=csv.reader(f)
header=next(data)

# row[0] 행정구역, row[1], row[2] : 총 인구수 
for row in data:
    if '5000000000' in row[0]:
        print(row)
f.close()
```

```python
f=open(d_pth)
data=csv.reader(f)
header=next(data)
ma=[]
fe=[]
for row in data:
    if '5000000000' in row[0]:
        for i in row[3:104]:
            ma.append(-int(i.replace(',','')))
        for i in row[106:]:
            fe.append(int(i.replace(',','')))

f.close()
plt.figure(figsize=(10,5))
plt.style.use('ggplot')
plt.barh(range(101),fe,color='r',label='female')
plt.barh(range(101),ma,color='b',label='male')
plt.legend()
plt.show()
```

## 혈액형 비율 표현하기

```python
plt.pie([10,20])
plt.show()
```

```python
size=[2441,2312,1031,1233]
plt.axis('equal')              # 동그랗게
plt.pie(size)
plt.show()
```

```python
# 레이블 표시
plt.rc('font',family='Malgun Gothic')
size=[2441,2312,1031,1233]
label=['A형','B형',' AB형','O형']
plt.axis('equal')              
plt.pie(size,labels=label)
plt.show()
```

```python
# 비율, 범례 표시
plt.rc('font',family='Malgun Gothic')
size=[2441,2312,1031,1233]
label=['A형','B형',' AB형','O형']
plt.axis('equal')              
plt.pie(size,labels=label,autopct='%.1f%%')      # autopct 비율 표시
plt.legend()                                     # 범례 표시
plt.show()
```

```python
# 효과
plt.rc('font',family='Malgun Gothic')
size=[2441,2312,1031,1233]
label=['A형','B형',' AB형','O형']
color=['darkmagenta','deeppink','hotpink','pink']    # 컬러 입력
plt.axis('equal')              
plt.pie(size,labels=label,autopct='%.1f%%',colors=color,explode=(0,0,0.1,0))# 세번째 ab형 돌출
plt.legend()                                    
plt.show()
```

## 제주도 성별 인구 비율 표현

```python
f=open(d_pth)
data=csv.reader(f)
header=next(data)
ma=0
fe=0
size=[]
for row in data:
    if '5000000000' in row[0]:
        for i in row[3:104]:
            ma+=int(i.replace(',',''))
        for i in row[106:]:
            fe+=int(i.replace(',',''))
size.append(ma)
size.append(fe)
f.close()
```

```python
plt.rc('font',family='Malgun Gothic')

label=['남성','여성']
color=['skyblue','pink']  
plt.axis('equal')              
plt.pie(size,labels=label,autopct='%.1f%%',colors=color,startangle=90)# 시작 각도 지정
plt.legend()                                    
plt.show()
```
