# 산점도로 시각화

---
## 꺾은선 그래프로 표현

```python
import csv
import matplotlib.pyplot as plt

d_pth='../data/gender.csv'
f=open(d_pth)
data=csv.reader(f)
ma=[]
fe=[]
for row in data:
    if '5000000000' in row[0]:
        for i in row[3:104]:
            ma.append(int(i.replace(',','')))
        for i in row[106:]:
            fe.append(int(i.replace(',','')))

f.close()
```

```python
plt.plot(ma,label='male')
plt.plot(fe,label='female')
plt.legend()
plt.show()
```

## 막대 그래프로 표현

```python
res=[]
for i in range(101):
    res.append(ma[i]-fe[i])
plt.bar(range(101),res)
plt.show()
# 60대 이상은 여성이 많다
```

## 버블차트로 표현

```python
plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,250,300],c=range(4),cmap='jet') #c=color
plt.colorbar()
plt.show()
```

```python
# 무작위로 100개의 점 만들기
import random
x=[]
y=[]
size=[]
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=.7) # alpha : 투명도
plt.colorbar()
plt.show()
```

## 제주도 연령대 별 성별 비율을 산점도로 표현

```python
f=open(d_pth)
data=csv.reader(f)
ma=[]
fe=[]
for row in data:
    if '5000000000' in row[0]:
        for i in row[3:104]:
            ma.append(int(i.replace(',','')))
        for i in row[106:]:
            fe.append(int(i.replace(',','')))

f.close()
```

```python
plt.scatter(ma,fe,c=range(101),alpha=.5,cmap='jet')
plt.colorbar()
plt.xlabel('male')
plt.ylabel('female')
plt.plot(range(max(ma)),range(max(ma)),'g')       # 추세선
plt.show()
```
