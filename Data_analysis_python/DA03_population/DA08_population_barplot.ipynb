## 동네를 입력받아 그래프 그리기

```python
d_pth='../data/age.csv'

import csv
import matplotlib.pyplot as plt

f=open(d_pth)
data=csv.reader(f)
header=next(data)
res=[]
name=input('동네를 입력하세요 : ')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            res.append(int(i))
f.close()

plt.style.use('ggplot')
plt.plot(res)
plt.show()
```

## bar(), barh()

```python
plt.bar([0,1,2,4,6,10],[1,2,3,5,6,7])    # bar([막대 표시 위치],[막대 높이])
plt.show()
```

```python
# range 사용
plt.bar(range(6),[1,2,3,5,6,7])
plt.show()
```

```python
# 구의동 인구 막대그래프로 표시하기
f=open(d_pth)
data=csv.reader(f)
header=next(data)
guui=[]
for row in data:
    if '구의제1동' in row[0]:
        for i in row[3:]:
            guui.append(int(i))
f.close()

plt.figure(figsize=(10,5))
plt.bar(range(101),guui)
plt.show()
```

```python
# barh()
plt.barh(range(101),guui)
plt.show()
```

## 남녀 인구 분포 시각화

[Data](https://jumin.mois.go.kr/index.jsp#) > 연령별 인구현황 > 통계표    
2019.02, 남녀 구분 O, 연령 구분 단위는 1세    
만 연령구분 : 0, 100 이상    
전체읍면동현황에 체크하여 csv 파일 다운로드 

```python
d_pth='../data/gender.csv'
```

```python
import csv

f=open(d_pth)
data=csv.reader(f)
header=next(data)
n=0
print(header)

for row in data:
    print(row)
    n+=1
    if n==2:
        print('...중략...')
        break
f.close()
```

```python
# 남성과 여성을 각각 다른 변수에 저장
f=open(d_pth)
data=csv.reader(f)
header=next(data)
ma=[]
fe=[]

for row in data:
    if '구의제1동' in row[0]:
        for i in range(101):
            ma.append(int(row[i+3]))
            fe.append(int(row[-(i+1)]))
fe.reverse()
f.close()
```

```python
plt.figure(figsize=(10,5))
plt.barh(range(101),fe,label='female')
plt.barh(range(101),ma,label='male')
plt.legend()
plt.show()
```

```python
# barh를 겹치지 않게 표현
new_ma=[-x for x in ma]

plt.figure(figsize=(10,5))
plt.barh(range(101),fe,label='female')
plt.barh(range(101),new_ma,label='male')
plt.legend()
plt.show()
```

```python
# 동네를 입력받아서 그래프 그리기
f=open(d_pth)
data=csv.reader(f)
header=next(data)
ma=[]
fe=[]
name=input('동네를 입력하세요 : ')
for row in data:
    if name in row[0]:
        for i in range(101):
            ma.append(int(row[i+3]))
            fe.append(int(row[-(i+1)]))
            
new_ma=[-x for x in ma]            
fe.reverse()
f.close()
plt.figure(figsize=(10,5))
plt.barh(range(101),fe,label='female')
plt.barh(range(101),new_ma,label='male')
plt.legend()
plt.show()
```
