# k-means clustering
---
### 1. k-means clustering

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

sns.set()
```

```python
x,yt=make_blobs(n_samples=300,centers=4,cluster_std=.6,random_state=0)
plt.scatter(x[:,0],x[:,1],s=50)
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(x)
ykmeans=kmeans.predict(x)
# print(ykmeans)
plt.scatter(x[:,0],x[:,1],c=ykmeans,s=50,cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=.5)
```

### 2. k-means clustering  

```python
from sklearn.datasets import make_moons
x,y=make_moons(200,noise=.05,random_state=0)

labels=KMeans(2,random_state=0).fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=labels,s=50,cmap='viridis')
```

### 3. k-means clustering digits dataset

```python
from sklearn.datasets import load_digits
digits=load_digits()
digits.data.shape
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=10,random_state=0)
clusters=kmeans.fit_predict(digits.data)

print(kmeans.cluster_centers_.shape)

fig,ax=plt.subplots(2,5,figsize=(8,3))
centers=kmeans.cluster_centers_.reshape(10,8,8)
for axi,center in zip(ax.flat,centers):
    axi.set(xticks=[],yticks=[])
    axi.imshow(center, interpolation='nearest',cmap=plt.cm.binary)
```

### 4. k-means clustering iris dataset

```python
from sklearn.datasets import load_iris 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import math 
%matplotlib inline 

iris = load_iris() 
iris_data = iris.data 
iris_data_pd = pd.DataFrame(iris_data, columns = iris.feature_names)
print (iris_data_pd)
```

```python
petals = pd.DataFrame(iris_data_pd.values[:,2:4], columns = ['petal length (cm)','petal width (cm)']) 
plt.scatter(petals.values[:,0],petals.values[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans 
plt.figure(figsize=(7, 5)) 
km = KMeans(n_clusters= 2, random_state=20)
km.fit(iris_data_pd.iloc[:,2:4]) 
y_pred = km.predict(iris_data_pd.iloc[:,2:4])
plt.scatter(iris_data_pd.iloc[:,2],iris_data_pd.iloc[:,3], c=y_pred)
plt.title("Clustering") 
plt.xlabel('petal length') 
plt.ylabel('petal width') 
plt.show()
```

```python
print (km.predict(iris_data_pd.iloc[:,2:4]))
```

```python
print(iris_data_pd.iloc[98,2:4])
```

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1 
    dy = y2 - y1 
    squared = dx**2 + dy**2
    result = math.sqrt(squared)
    return result 

print ('0 cluster distance: ', distance(iris_data_pd.iloc[98,2], iris_data_pd.iloc[98,3], km.cluster_centers_[0][0], 
km.cluster_centers_[0][1])) 
print ('1 cluster distance: ', distance(iris_data_pd.iloc[98,2], iris_data_pd.iloc[98,3], km.cluster_centers_[1][0], 
km.cluster_centers_[1][1]))
```

```python
n_cluster = [3,4,6,12] 
for i in n_cluster:
    count = 1 
    km = KMeans(n_clusters= i, random_state=20) 
    km.fit(iris_data_pd.iloc[:,2:4]) 
    y_pred = km.predict(iris_data_pd.iloc[:,2:4]) 
    plt.figure(count) 
    plt.scatter(iris_data_pd.iloc[:,2],iris_data_pd.iloc[:,3], c=y_pred)
    plt.title("Clustering = "+ str(i)) 
    plt.xlabel('petal length') 
    plt.ylabel('petal width') 
    count=count+1 
    plt.show()
```
