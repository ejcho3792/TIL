# anscombe dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
anscombe=sns.load_dataset('anscombe')
anscombe.head()
```

```python
data1=anscombe[anscombe['dataset']=='I']
data2=anscombe[anscombe['dataset']=='II']
data3=anscombe[anscombe['dataset']=='III']
data4=anscombe[anscombe['dataset']=='IV']

fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.set_title('data1')
ax2.set_title('data2')
ax3.set_title('data3')
ax4.set_title('data4')

ax1.plot(data1['x'],data1['y'],'o')
ax2.plot(data2['x'],data2['y'],'o')
ax3.plot(data3['x'],data3['y'],'o')
ax4.plot(data4['x'],data4['y'],'o')

fig.suptitle('anscombe data')
fig.tight_layout( )
```

# Tips dataset
using matplotlib

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
ax1.hist(tips['total_bill'],bins=10)
ax1.set_title('histogram of total bill')
ax1.set_xlabel('frequency')
ax1.set_ylabel('total_bill')
```

```python
scatters=plt.figure()
ax1=scatters.add_subplot(1,1,1)
ax1.scatter(tips['total_bill'],tips['tip'])
ax1.set_title('scatterplot of total bill vs tip')
ax1.set_xlabel('total bill')
ax1.set_ylabel('tip')
```

```python
boxplot=plt.figure()
ax1=boxplot.add_subplot(1,1,1)
ax1.boxplot([tips[tips['sex']=='Female']['tip'],
            tips[tips['sex']=='Male']['tip']],
            labels=['Female','Male']
           )
ax1.set_title('boxplot of tips by sex')
ax1.set_xlabel('sex')
ax1.set_ylabel('tip')
```

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1

tips['sex_color']=tips['sex'].apply(recode_sex)

scatters=plt.figure()
ax1=scatters.add_subplot(1,1,1)
ax1.scatter(tips['total_bill'],tips['tip'],
            s=tips['size']*10,c=tips['sex_color'],alpha=0.5)
ax1.set_title('total bill vs tip colored by sex and sized by size')
ax1.set_xlabel('total bill')
ax1.set_ylabel('tip')
```

# Tips dataset
using seaborn

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
tips=sns.load_dataset('tips')
tips.head()
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'])
ax.set_title('total bill histogram with density plot')

ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title('total bill histogram')

ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False)
ax.set_title('total bill density')

ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title('total bill histogram with density and rug plot')

ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title('count of days')
```

```python
# 
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('scatterplot of total bill and tip')

ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips,fit_reg=False)
ax.set_title('scatterplot of total bill and tip')
```

```python
#jointplot
ax=sns.jointplot(x='total_bill',y='tip',data=tips)
ax.fig.suptitle('joint plot of total bill and tip',y=1.03)
```

```python
#jointplot hexbin
ax=sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
ax.fig.suptitle('joint plot of total bill and tip',y=1.03)
```

```python
#  kdeplot
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],data2=tips['tip'],shade=True)
ax.set_title('kernel density plot of total bill and tip')
```

```python
#bar graph barplot
ax=plt.subplots()
ax=sns.barplot(x='time',y='total_bill',data=tips)
ax.set_title('bar plot of average total bill for time of day')
```

```python
#box graph boxplot
ax=plt.subplots()
ax=sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('boxplot of total bill by time of day')

#violinplot
ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('violin plot of total bill by time of day')

ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
ax.set_title('violin plot of total bill by time of day')
```

```python
#pairplot
ax=sns.pairplot(tips)
```

```python
#pairplot hue=sex
ax=sns.pairplot(tips,hue='sex')
```

```python
#pairgrid
pair=sns.PairGrid(tips)
pair=pair.map_upper(sns.regplot)
pair=pair.map_lower(sns.kdeplot)
pair=pair.map_diag(sns.distplot,rug=True)
plt.show()
```

```python
#lmplot
ax=sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',fit_reg=False)
```
