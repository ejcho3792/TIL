# 전처리

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
happy=pd.read_excel('happyhappy.xlsx')
h=np.array(happy)[1,1:]
h
```

```python
d=np.zeros(len(h)+1)
for i in range(len(h)):
    d[i]=h[-(i+1)]
print(d)
```

```python
dd=pd.read_excel('analysis_data.xlsx')
dd
```

```python
dd['삶의 질']=d

dd.to_excel('analysis_data.xlsx',index=False)
dd
```

# 상관계수

```python
import pandas as pd

dd=pd.read_excel('analysis_data.xlsx')
dd=dd.set_index('분기')
dd
```

```python
ddd=dd.iloc[4:-1,:]
ddd
```

```python
ddd.corr()
```

```python
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(15,15))
sns.heatmap(data = ddd.corr(),vmin = -1, vmax = +1, annot=True,annot_kws={"size": 30} ,
fmt = '.2f', linewidths=.5, cmap='Blues')
sns.set(font_scale=3)
```
