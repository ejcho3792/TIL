```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d=pd.read_excel('문화활동수_자치구.xlsx')
d=d.set_index('분류',drop=False)
d
```

```python
dd=pd.read_excel('analysis_data_culture.xlsx')
dd=dd.set_index('연도')
dd
```

```python
gu=d['분류'].unique()
gu
```

```python
aa=[]
for i in gu:

    data=pd.DataFrame()
    data=d.loc[i]
    data=data.iloc[:,2:]
    data['국민행복지수']=np.array(dd['국민행복지수'])
    for j in data.columns:
        aa.append(data['국민행복지수'].corr(data[j]))
```

```python
co=pd.DataFrame(np.reshape(aa,(26,9)))
co.columns=['전시회 관람', '박물관 관람', '음악 및 무용발표회 관람', '전통예술공연 관람', '연극공연 관람',
       '극장에서 영화관람', '대중공연 관람', '스포츠경기 관람','국민행복지수']
co.index=d.index[:26]
co
```
