```python
data=dd[4:-1]
import numpy as np

data=np.array(data)
data=data[:,1:]
x,y=data[:,0],data[:,1]
```

```python
np.corrcoef(x,y)[0,1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.scatter(x,y)
plt.show()
colormap = plt.cm.PuBu

sns.heatmap(np.corrcoef(x,y), 
            linewidths = 0.1, 
            vmax = 1.0, 
            square = True, 
            cmap = colormap, 
            linecolor = "white", 
            annot = True, 
            annot_kws = {"size" : 16})

```

```python
dd=pd.read_csv('seoul_corona.csv',encoding='cp949')

ddd=dd['자치구 기준일']

for i in range(len(ddd)):
    
    if len(ddd[i])!=13:
        ddd[i]='20'+ddd[742]
for i in range(len(ddd)):
    ddd[i]= ddd[i][:4]+'-'+ ddd[i][5:7]+'-'+ddd[i][8:10]
#print(ddd)

dd['자치구 기준일']=ddd
dd
```

```python
dd['자치구 기준일']=pd.to_datetime(dd['자치구 기준일'])

dd['분기']=dd['자치구 기준일'].dt.to_period('Q').dt.strftime('%Y %q분기')
dd
```
