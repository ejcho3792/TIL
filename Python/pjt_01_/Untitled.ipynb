```python
dd=pd.read_csv('seoul_corona.csv',encoding='cp949')

ddd=dd['자치구 기준일']

for i in range(len(ddd)):
    
    if len(ddd[i])!=13:
        ddd[i]='20'+ddd[742]
for i in range(len(ddd)):
    ddd[i]= ddd[i][:4]+'-'+ ddd[i][5:7]+'-'+ddd[i][8:10]
#print(ddd)

dd['자치구 기준일']=ddd
dd
```

```python
dd['자치구 기준일']=pd.to_datetime(dd['자치구 기준일'])

dd['분기']=dd['자치구 기준일'].dt.to_period('Q').dt.strftime('%Y %q분기')
dd
```
