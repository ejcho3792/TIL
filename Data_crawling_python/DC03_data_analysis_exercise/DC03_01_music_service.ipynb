```python
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

---
## 1. 멜론차트 크롤링하여 엑셀로 저장

```python
service=Service('../chromedriver/chromedriver.exe')
driver=webdriver.Chrome(service=service)
```

```python
url='https://www.melon.com/chart/'
driver.get(url)

html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
songs=soup.select('table>tbody>tr')
print('title',songs[0].select('div.ellipsis.rank01>span>a')[0].text)
print('singer',songs[0].select('div.ellipsis.rank02>span>a')[0].text)
```

```python
song_data=[]
rank=1
songs=soup.select('table>tbody>tr')
for song in songs:
    title=song.select('div.ellipsis.rank01>span>a')[0].text
    singer=song.select('div.ellipsis.rank02>span>a')[0].text
    song_data.append(['Melon',rank,title,singer])
    rank+=1
```

```python
columns=['service','rank','title','singer']
pd_data=pd.DataFrame(song_data,columns=columns)
pd_data
```

```python
pd_data.to_excel('./files/my_melon.xlsx',index=False)
```

## 2. 벅스 크롤링하여 엑셀로 저장

```python
service=Service('../chromedriver/chromedriver.exe')
driver=webdriver.Chrome(service=service)
```

```python
url='https://music.bugs.co.kr/chart'
driver.get(url)

html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
songs=soup.select('table.byChart>tbody>tr')
print(len(songs))
print(songs[0].select('p.title>a')[0].text)
print(songs[0].select('p.artist>a')[0].text)
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select('p.title>a')[0].text
    singer=song.select('p.artist>a')[0].text
    song_data.append(['Bugs',rank,title,singer])
    rank+=1
```

```python
columns=['service','rank','title','singer']
pd_data=pd.DataFrame(song_data,columns=columns)
pd_data
```

```python
pd_data.to_excel('./files/my_bugs.xlsx',index=False)
```

## 3. 지니 크롤링하여 엑셀로 저장

```python
service=Service('../chromedriver/chromedriver.exe')
driver=webdriver.Chrome(service=service)
```

```python
url='https://mw.genie.co.kr/chart'
driver.get(url)

html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
songs=soup.select('div.info>a')
print(len(songs))
print(songs[0].select('p')[0].text.strip())
print(songs[0].select('span')[0].text)
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select('p')[0].text.strip()
    singer=song.select('span')[0].text
    song_data.append(['Genie',rank,title,singer])
    rank+=1
```

```python
columns=['service','rank','title','singer']
pd_data=pd.DataFrame(song_data,columns=columns)
pd_data
```

```python
pd_data.to_excel('./files/my_genie.xlsx',index=False)
```
